```python
from pathlib import Path
```

```python
text = Path("test.md").read_text()
```

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter, RecursiveCharacterTextSplitter

headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
]

# MD splits
markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on, strip_headers=False
)

md_header_splits = markdown_splitter.split_text(text)

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512, chunk_overlap=200, separators=["\n\n", "\n", " ", "", "</TABLE>"]
)

# Split
splits = text_splitter.split_documents(md_header_splits)
```

```python
from termcolor import colored
from itertools import cycle

def print_chunks(chunks: list):
    for chunk, color in zip(chunks, cycle(["red", "blue", "green"])):
        chunk_text = colored(chunk.page_content, color, attrs=["reverse", "blink"])
        print(chunk_text)
```

```python
print_chunks(splits)
```
